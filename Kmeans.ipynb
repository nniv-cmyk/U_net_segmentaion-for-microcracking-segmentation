```python
import numpy as np
import cv2

# Load an image
img = cv2.imread("image/C_6.jpg")
```

```python
# Convert the image to a matrix of pixel values
Z = img.reshape((-1,3))

# Convert the matrix to float32
Z = np.float32(Z)
```

```python
# Define the number of clusters (K)
K = 3

# Apply K-means clustering
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
# Convert the segmented regions back into an image format
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((img.shape))

# Display the segmented image
# cv2.imshow("Segmented Image", res2)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Display the first image in the first subplot
ax1.imshow(img[:,:,::-1])
# ax1.axis('off')

# Display the second image in the second subplot
ax2.imshow(res2[:,:,::-1])
# ax2.axis('off')

# Show the figure
plt.show()
```

```python
# Define the crop coordinates
x, y, w, h = 1000, 0, 1500, 5000

# Crop the image using NumPy slicing
cropped = img[y:y+h, x:x+w].copy()
```

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Display the first image in the first subplot
ax1.imshow(img[:,:,::-1])
# ax1.axis('off')

# Display the second image in the second subplot
ax2.imshow(cropped[:,:,::-1])
# ax2.axis('off')

# Show the figure
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["figure.figsize"] = (10,10)
matplotlib.rcParams["image.cmap"] = "gray"
```

```python
img = cv2.imread("image/color.jpg")
```

```python
img.shape
```

```python
plt.imshow(img[:,:,::-1])
```

```python
imgB,imgG,imgR = cv2.split(img)
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray)
plt.title("Grayscale")
```

```python
plt.figure(figsize=(20,10))
plt.subplot(151)
plt.imshow(img[:,:,::-1])
plt.title("Original")
plt.subplot(152)
plt.imshow(imgGray)
plt.title("Grayscale")
plt.subplot(153)
plt.imshow(imgB)
plt.title("Blue")
plt.subplot(154)
plt.imshow(imgG)
plt.title("Green")
plt.subplot(155)
plt.imshow(imgR)
plt.title("Red")
```

```python
# Hue - indicates the color / tint of the pixel

# Saturation - indicates the purity (or richness) of the color

# Value - indicates the amount of brightness of the pixel
```

```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)
```

```python
# convert from bgr to HSV format
hsvImage = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(20,15))
plt.subplot(131)
plt.imshow(hsvImage[:,:,0],cmap='gray')
plt.title("Hue");
plt.subplot(132)
plt.imshow(hsvImage[:,:,1],cmap='gray')
plt.title("Saturation");
plt.subplot(133)
plt.imshow(hsvImage[:,:,2],cmap='gray')
plt.title("Value");
```

```python
H, S, V = cv2.split(hsvImage)
print(H.shape)
```

```python
# Remove unsaturated (white/gray) pixels 
H_array = H[S > 10].flatten()
H_array.shape
```

```python
plt.figure(figsize=[20,10])
plt.subplot(121);plt.imshow(img[...,::-1]);plt.title("Image");plt.axis('off')
plt.subplot(122);plt.hist(H_array, bins=180, color='r');plt.title("Histogram")
```
